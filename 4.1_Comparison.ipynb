#### Reclassify Categories Unique To GL

```python
df = gl_summary.copy()

grp_Deps = df.iloc[1:3].sum()
grp_Checks = df.iloc[3:5].sum()
grp_OtherPay = df.iloc[5:8].sum()
grp_CLa = df.iloc[8]
grp_CLp = df.iloc[9]
grp_PRx = df.iloc[10]

gl_alt_summary = pd.concat([grp_Deps,grp_Checks,grp_OtherPay,
                            grp_CLa,grp_CLp,grp_PRx],axis=1).T

gl_alt_summary = gl_alt_summary.rename(index={0: 'Credits', 1: 'Checks', 2:'Other_Pay'})

bank_summary = b_summary.drop(['Beg', 'End']).rename(index={'Deposits': 'Credits'})
```

<br>

#### Calculate Variances: GL - Bank

```python
variances = gl_alt_summary - bank_summary.values
variances.style.format('{0:,.0f}')
```

<br> 
<br> 

#### Summary Stats


```python
stats = pd.concat([bank_summary,gl_alt_summary], keys=['bank', 'gl']) \
           .stack() \
           .unstack(0) \
           .swaplevel() \
           .rename_axis(['month','type'])
```
